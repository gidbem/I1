```python
import pandas as pd
import numpy as np
```

```python
dados_catalogo = {
'ID_Titulo': [101, 102, 103, 104, 105, 106, 101, 107, 108],
'Titulo': ['Stranger Things', 'O Poderoso Chefão', 'Interestelar', 'Breaking Bad', 'Duna', 'Succession', 'Stranger Things', 'Toy Story', 'Vingadores: Ultimato'],
'Tipo': ['Série', 'Filme', 'Filme', 'Série', 'Filme', 'Série', 'Série', 'Filme', 'Filme'],
'Ano': [2016, 1972, np.nan, 2008, 2021, 2018, 2016, 1995, 2019],
'Genero': ['Ficção Científica', np.nan, 'Ficção Científica', 'Drama', 'Ficção Científica', 'Drama', 'Ficção Científica', 'Animação', 'Ação'],
'Plataforma': ['Netflix', 'Prime Video', 'Prime Video', 'Netflix', 'HBO Max', 'HBO Max', 'Netflix', 'Disney+', 'Disney+']
}
df_catalogo = pd.DataFrame(dados_catalogo)

display(df_catalogo)
```

```python
dados_metricas = {
'ID_Titulo': [101, 102, 103, 104, 105, 106, 107, 109], # O ID 109 é de um filme que não está no catálogo
'Nota_IMDb': [8.7, 9.2, 8.6, 9.5, 8.0, 8.8, 8.3, 9.0],
'Votos_Milhares': [1200, 1900, 1500, 1800, 700, 450, 600, 950],
'Orcamento_Milhoes': [12.0, 6.0, 165.0, 3.0, 165.0, 9.0, 30.0, 180.0]
}
df_metricas = pd.DataFrame(dados_metricas)
print("Tabelas carregadas com sucesso! Pronto para iniciar a análise.")

display(df_metricas)
```

```python
#FASE 01
#1
print(df_catalogo.duplicated().sum()) 
df_catalogo = df_catalogo.drop_duplicates()
display(df_catalogo)
```

```python
#2 
df_catalogo['Genero'] = df_catalogo['Genero'].fillna('Clássico')
display(df_catalogo)
```

```python
#3
df_catalogo['Ano'] = df_catalogo['Ano'].fillna(2014)
display(df_catalogo)
```

```python
#4
print(df_catalogo.isnull().sum())
```

```python
#FASE 02
#1
print(df_catalogo.loc[:2,['Titulo', 'Plataforma']])
print()
print()
print(df_catalogo.iloc[4, 1])
```

```python
#2
print(df_catalogo[(df_catalogo['Genero'] == 'Ficção Científica') & (df_catalogo['Ano'] >= 2015)])
```

```python
#3
def categoria_orcamento(n):
    if n < 10:
        return 'Baixo Custo'
    
    elif n >= 10 and n < 100:
        return 'Médio Custo'
    
    elif n >= 100:
        return 'Blockbuster'
```

```python
#FASE 04
#1
df_consolidado = pd.merge(df_catalogo, df_metricas, on='ID_Titulo', how='left')
display(df_consolidado)
```

```python
#2
df_consolidado['Nota_IMDb'] = df_consolidado['Nota_IMDb'].fillna(df_consolidado['Nota_IMDb'].mean())
df_consolidado['Votos_Milhares'] = df_consolidado['Votos_Milhares'].fillna(df_consolidado['Votos_Milhares'].mean())
df_consolidado['Orcamento_Milhoes'] = df_consolidado['Orcamento_Milhoes'].fillna(0.0)
display(df_consolidado)

```

```python
#3
print(df_consolidado.groupby('Genero').count())
```

```python
#4
soma_total_orcamento = df_consolidado.groupby('Genero')['Orcamento_Milhoes'].sum()
print('Soma total dos orçamentos: ')
print()
print(soma_total_orcamento)
print()
media_notas_genero = df_consolidado.groupby('Genero')['Nota_IMDb'].mean()
print('Média das notas: ')
print()
print(media_notas_genero)
print()
print("Gênero com maior orçamento:", soma_total_orcamento.idxmax())
print("Gênero com melhor nota média:", media_notas_genero.idxmax())
```

```python
#5
df_consolidado['Perfil_Financeiro'] = df_consolidado['Orcamento_Milhoes'].apply(categoria_orcamento)
display(df_consolidado)
```
